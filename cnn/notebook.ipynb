```python
model = Sequential([
   Input(shape = (20, 20, 3)),
   Conv2D(filters = 32, kernel_size = (3,3), padding = 'same', activation = 'relu' ),
   Conv2D(filters = 32, kernel_size = (3,3), padding = 'same', activation = 'relu' ),
   MaxPooling2D(),
   Conv2D(filters = 16, kernel_size = (3,3), padding = 'same', activation = 'relu' ),
   Conv2D(filters = 16, kernel_size = (3,3), padding = 'same', activation = 'relu' ),
   MaxPooling2D(),
   Conv2D(filters = 8, kernel_size = (3,3), padding = 'same', activation = 'relu' ),
   Conv2D(filters = 8, kernel_size = (3,3), padding = 'same', activation = 'relu' ),
   MaxPooling2D(),
   Flatten(),
   Dense(24,activation = 'relu'),
   Dense(10,activation = 'softmax')
])
```

```python
BASEDIR = "C:/Users/as/Desktop/Deep Learning/Deep-Learning/data/natural_images/"
classes = os.listdir(BASEDIR)
folder_names = [os.path.join(BASEDIR,class_name) for class_name in classes]
```

```python
all_files = []
all_classes = []


for folder_name,class_name in zip(folder_names,classes):
    files = os.listdir(folder_name)
    all_files.extend(os.path.join(folder_name, file) for file in files)
    all_classes.extend([class_name]*len(files))
```
